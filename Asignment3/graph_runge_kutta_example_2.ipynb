```python
import numpy as np
import matplotlib.pyplot as plt

# Define the ODE function f(y, t)
def f(y, t):
    return y * t**2

# Define the analytical solution y(t)
def y(t):
    return -2 / (t**2 - 2)

def runge_kutta_4th_order(y0, t0, tau):
    k1 = f(y0, t0)
    k2 = f(y0 + tau * k1 / 2.0, t0 + tau / 2.0)
    k3 = f(y0 + tau * k2 / 2.0, t0 + tau / 2.0)
    k4 = f(y0 + tau * k3, t0 + tau)

    y1 = y0 + tau * (k1 + 2 * k2 + 2 * k3 + k4) / 6.0
    return y1

t0 = 0.0
t_end = 5
num_intervals = 100
tau = (t_end - t0) / num_intervals

time_points = []
y_values = []

for i in range(num_intervals):
    y0 = y(t0 + i * tau) 
    y1 = runge_kutta_4th_order(y0, t0 + i * tau, tau)
    time_points.append(t0 + i * tau)
    y_values.append(y1)

# Plot the numerical solution
time_exact = np.linspace(t0, t_end, num_intervals*10)
y_exact = y(time_exact)
nan_thresh = 200
y_exact[y_exact < -nan_thresh] = np.nan
y_exact[y_exact > nan_thresh] = np.nan
fig = plt.figure(figsize=(14, 8))
plt.scatter(time_points, y_values, label='Numerical Solution (4th Order RK)', color='r', marker='o')
plt.plot(time_exact, y_exact, color='green', label='Analytic solution (y = -2 / (t**2 - 2)')
plt.xlabel('t')
plt.ylabel('y')
plt.legend()
plt.show()
```
